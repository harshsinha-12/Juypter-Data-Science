```python
roko = pd.read_csv("/Users/harsh/VS Code Codes/Juypter Data Science/Datasets/datasets-session-23/sharma-kohli.csv")
```

```python
roko.head(4)
```

```python
roko.rename(columns={'index': 'Year', 'RG Sharma': 'Rohit','V Kohli': 'Virat',
                    }, inplace=True)
```

```python
plt.plot(roko["Year"], roko["Virat"], label = "Kohli")
plt.plot(roko["Year"], roko["Rohit"], label = "Rohit")
```

```python
plt.plot(roko["Year"], roko["Virat"], label = "Kohli")
plt.plot(roko["Year"], roko["Rohit"], label = "Rohit")

plt.title("Kohli vs Rohit IPL Season Performance")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
```

```python
plt.plot(roko["Year"], roko["Virat"], label = "Kohli", color = "red", marker = "o")
plt.plot(roko["Year"], roko["Rohit"], label = "Rohit", color = "blue", marker = "o")

plt.title("Kohli vs Rohit IPL Season Performance")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
```

```python
plt.plot(roko["Year"], roko["Virat"], label = "Kohli", 
        color = "red", marker = "o", linestyle = "--", linewidth = 3)
plt.plot(roko["Year"], roko["Rohit"], label = "Rohit", 
        color = "blue", marker = "o", linestyle = "-.")

plt.title("Kohli vs Rohit IPL Season Performance")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend(loc = "upper left")
```
