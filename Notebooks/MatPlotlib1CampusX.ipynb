```python
# Stacked Bar
plt.bar(df['batsman'], df['2015'], color = 'red', label = '2017')
plt.bar(df['batsman'], df['2016'], color = 'blue', bottom = df['2015'], label = '2016')
plt.bar(df['batsman'], df['2017'], color = 'green', bottom = df['2015'] + df['2016'], label = '2015')
plt.legend()
```

# Pie Chart
1. Univariate/Bivariate
2. Categorical 
3. %tage contribution

```python
data = [23,45,100,20,49]
subjects = ['english','maths','science','hindi','social']
plt.pie(data, labels = subjects)
plt.show()
```
