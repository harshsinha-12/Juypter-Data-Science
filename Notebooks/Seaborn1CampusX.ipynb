```python
# Matrix Plot -> heatmap, clustermap
```

```python
gap.sample(5)
```

```python
temp = gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
```

```python
temp.sample(5)
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(data = temp, cmap = 'winter')
```

```python
temp = gap[gap['continent'] == 'Asia'].pivot(index='country', 
                                            columns='year', values='lifeExp')

```

```python
plt.figure(figsize = (15,15))
sns.heatmap(data = temp, cmap = 'winter', annot=True, fmt = '.1f', linewidths = 1)
```
