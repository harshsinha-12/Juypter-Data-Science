# Annotation

```python
batters = pd.read_csv("/Users/harsh/VS Code Codes/Juypter Data Science/Datasets/datasets-session-24/batter.csv")
```

```python
sample = batters.head(100).sample(25, random_state = 5)
```

```python
sample.head()
```

```python
plt.figure(figsize = (11, 7))
plt.scatter(sample['avg'], sample['strike_rate'], c = sample['runs'], s= 300, cmap = 'winter')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
colorbar = plt.colorbar()
colorbar.set_label('Runs')
for i in range(len(sample)):
    plt.text(sample['avg'].values[i], sample['strike_rate'].values[i], sample['batter'].values[i])
```

# Horizontal and Vertical Lines

```python
plt.figure(figsize = (11, 7))
plt.scatter(sample['avg'], sample['strike_rate'], c = sample['runs'], s= 300, cmap = 'winter')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.axhline(130,color = 'red', linestyle = '--')
plt.axvline(30,color = 'red', linestyle = '--')
plt.axhline(140,color = 'green', linestyle = '--')
colorbar = plt.colorbar()
colorbar.set_label('Runs')
plt.title('Batsmen Analysis in IPL')
for i in range(len(sample)):
    plt.text(sample['avg'].values[i], sample['strike_rate'].values[i], sample['batter'].values[i])
```

# 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize = (10, 7))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap = 'winter')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (10, 7))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap = 'winter')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)

fig = plt.figure(figsize = (10, 7))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap = 'winter')
fig.colorbar(p)
```

# Contour Plot

```python

z = xx**2 + yy**2
fig = plt.figure(figsize = (10, 7))
ax = plt.subplot()

p = ax.contour(xx,yy,z, cmap = 'winter')
fig.colorbar(p)
```

```python
z = xx**2 + yy**2
fig = plt.figure(figsize = (10, 7))
ax = plt.subplot()

p = ax.contourf(xx,yy,z, cmap = 'winter')
fig.colorbar(p)
```
